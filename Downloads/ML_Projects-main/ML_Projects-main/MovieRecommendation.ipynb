# **Preprocessing of the Data**

```python
# Keep the necessary coloumns
movies= movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
#check for the missing values
movies.isnull().sum()
```

```python
#Drop the rows which have null values
movies.dropna()
```

```python
#Check for the duplicate values
movies.duplicated().sum()
```

```python
#Check the genre coloumn format
movies.iloc[0].genres
```

```python
#Loading the library to convert the string os list into the list
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
#Function to get the genres
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
      L.append(i['name'])
  return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#Function for Taking the first 3 names from the cast
def convert3(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
      if counter != 3:
          L.append(i['name'])
          counter+=1
      else:
        break
  return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head(5)
```

```python
#Function To get the director name from the entire crew list
def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head(5)
```

```python
# Convert the overview which is a string into the list
movies['overview'][0]
```

```python
#Some values are float aswell
#astype(str) function is applied to convert any non string values to string
#split() function is applied on the string values
movies['overview']=movies['overview'].fillna('').astype(str).apply(lambda x: x.split())
```

```python
movies.head(3)
```

```python
#Removing the gaps in the coloumns
movies['genres']=movies['genres'].apply(lambda x:[i.replace( " ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace( " ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace( " ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace( " ","") for i in x])
```

```python
movies.head(3)
```

```python
movies['tags']=movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(3)
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head(3)
```

```python
#Adding spaces in the tags coloumn
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df
```

```python
#convert the tags column into lower case
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df
```

```python
#Loading the nltk library and porterstemmer to stem the values
import nltk
```

```python
from nltk.stem import PorterStemmer

def stem(text):
    ps = PorterStemmer()
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)

```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#Performing vectorization to convert texts to vectors in the tags coloumn
#using the countvectorizer class from sklearn
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
# Finding the cosine distance between the two vectors.(Lesser the distance more the similarities)
# Done using the function from sckitlearn
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Calculating the distance between one vector to every other vector
similarity = cosine_similarity(vectors)
```

```python
similarity[0]  # checks the similarity of the first vector with every other vector I.e 4806 values
```

```python
#Function to recommend a movie
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
